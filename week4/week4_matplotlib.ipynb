```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data,bins=30)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes=[30,25,15,10,5,5]

plt.pie(sizes)
plt.title("Pie Chart")
plt.show()
```
